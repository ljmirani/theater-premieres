#### Scraping the dates

```python
date_entries = soup.find_all('div', class_='tile-item-image-overlay--title has-inverted-text')
```

```python
dates = []
for entry in date_entries:
    date_tag = entry.find('h4', class_='is-headline-sub is-secondary')
    title_tag = entry.find('h3', class_='h4')
    
    date = date_tag.get_text(strip=True) if date_tag else 'no date'
    dates.append(date)
    print(f'Datum: {date}')
    print('---')
```

#### Scraping the descriptions

```python
def get_description(url):
        response = requests.get(url)
        response.raise_for_status()
      
        soup = BeautifulSoup(response.text, 'html.parser')
        
        div_content = soup.find('div', class_='description-wrapper')

        if not div_content:
            print(f"not found: {url}")
            return None

        paragraphs = div_content.find_all('p')
 
        description = "\n\n".join(p.get_text() for p in paragraphs)
        
        return description
        
    
```

```python
descriptions = []
for link in list_links:
    description = get_description(link)
    print("Description:", description)
    descriptions.append(description)
    print("----")
```

```python
len(descriptions)
```

#### Scraping the director

```python

def get_director(url):
    response = requests.get(url)
    response.raise_for_status()
      
    soup = BeautifulSoup(response.text, 'html.parser')
    
   
    div_contents = soup.find_all('div', {'class': 'field-collection-item field-collection-item--name-field-cast field-collection-item--view-mode-full tile-item--cast-item'})
    
    for div_content in div_contents:
        div_cast = div_content.find('div', {'class': 'tile-item--cast-item-cast'})

        span_content = div_cast.find('span', {'class': 'is-headline-sub is-secondary'})
    
        if 'Regie' in span_content.get_text():
            a_contents = div_content.find_all('a', {'class': 'tile-item--cast-item-person h4'})
     
            regie_names = [a.get_text().strip() for a in a_contents]
            
            return regie_names
    
    return None
```

```python
director_names = []
for link in list_links:
    director_name = get_director(link)
    if director_name is not None:
        director_name = ', '.join(director_name)
    director_names.append(director_name)

    print(director_name)
    print("----")
```
